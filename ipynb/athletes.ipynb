## Athlete Name and UFC Page

```python
# Read CSV with data
df = pd.read_csv('../csv/combined_fighter.csv')
```

```python
# Isolate athlete names
# Drop duplicates
athletes_df = pd.DataFrame(df["fighter"].drop_duplicates())
athletes_df
```

```python
# Split fighter name to first and last name

splitted = athletes_df['fighter'].str.split()
athletes_df['first_name'] = splitted.str[0]
athletes_df['last_name'] = splitted.str[-1]
middle = splitted.str[1] 
athletes_df['middle_name'] = middle.mask(middle == athletes_df['last_name'], '')
athletes_df['webpage'] = "https://www.ufc.com/athlete/" + athletes_df['first_name'] + "-" + athletes_df['last_name']

athletes_df
```

```python
#FIX MIDDLE NAME ISSUE
#COUNT HOW MANY VALUES HAVE A MIDDLE NAME (either manually fix

# middlename = athletes_df['middle_name'] != ""
# middlename

import numpy as np

athletes_df['middle_name'].replace('', np.nan, inplace = True)
# athletes_df

athletes_df.dropna(subset = ['middle_name'], inplace = True)
athletes_df.to_csv('athletes_df.csv')
```
